# Orbit Initialization

This tutorial covers all the ways to initialize orbits in galpy, from
raw phase-space coordinates to observed sky positions and astropy SkyCoord objects.

```python
%matplotlib inline
import numpy
from astropy import units
from galpy.orbit import Orbit
from galpy.potential import MWPotential2014
```

## Standard 3D initialization

The most basic way to create an orbit is by specifying
phase-space coordinates in galactocentric cylindrical coordinates
`[R, vR, vT, z, vz, phi]` (in natural units where $R_0 = 1$, $V_0 = 1$).

```python
# Full 3D orbit with azimuth
o = Orbit([1.0, 0.1, 1.0, 0.0, 0.05, 0.0])
print(o)
print("R =", o.R(), "  vT =", o.vT())
```

If the orbit is in an axisymmetric potential, the azimuth `phi` can be omitted:

```python
# 3D orbit without azimuth (axisymmetric potentials)
o_noaz = Orbit([1.0, 0.1, 1.0, 0.0, 0.05])
print(o_noaz)
```

## 2D and 1D orbits

galpy also supports planar (2D) and linear (1D) orbits.

```python
# 2D orbit with azimuth: [R, vR, vT, phi]
o2d = Orbit([1.0, 0.1, 1.0, 0.0])
print("2D with phi:", o2d)

# 2D orbit without azimuth: [R, vR, vT]
o2d_noaz = Orbit([1.0, 0.1, 1.0])
print("2D no phi:", o2d_noaz)

# 1D orbit: [x, vx]
o1d = Orbit([1.0, 0.1])
print("1D:", o1d)
```

## Physical units with `ro=` and `vo=`

By setting `ro` (distance scale in kpc) and `vo` (velocity scale in km/s),
the orbit will return quantities in physical units.

```python
o_phys = Orbit([1.0, 0.1, 1.0, 0.0, 0.05, 0.0], ro=8.0, vo=220.0)
print("R =", o_phys.R(), "kpc")
print("vT =", o_phys.vT(), "km/s")
```

## Astropy Quantity inputs

You can also initialize with astropy Quantities directly.

<div class="alert alert-info">

**Tip:** If you do input and output in physical units, the internal unit conversion
specified by `ro=` and `vo=` does not matter!

</div>

```python
o_qty = Orbit(
    [
        8.0 * units.kpc,
        22.0 * units.km / units.s,
        220.0 * units.km / units.s,
        0.0 * units.kpc,
        11.0 * units.km / units.s,
        0.0 * units.rad,
    ]
)
print("R =", o_qty.R(), "kpc")
print("vT =", o_qty.vT(), "km/s")
```

## From observed coordinates

Initialize from sky coordinates using `radec=True` (RA/Dec in degrees,
distance in kpc, proper motions in mas/yr, radial velocity in km/s)
or `lb=True` (Galactic l,b).

```python
# From RA, Dec: [RA, Dec, dist, pmRA*cos(Dec), pmDec, vlos]
o_radec = Orbit(
    [20.0, 30.0, 2.0, 3.5, -1.2, 15.0],
    radec=True,
    ro=8.0,
    vo=220.0,
)
print("RA =", o_radec.ra(), "deg")
print("Dec =", o_radec.dec(), "deg")
```

Similarly, from Galactic coordinates `(l, b)`:

```python
# From Galactic l, b
o_lb = Orbit(
    [150.0, 40.0, 1.5, 2.0, -0.5, 30.0],
    lb=True,
    ro=8.0,
    vo=220.0,
)
print("l =", o_lb.ll(), "deg")
print("b =", o_lb.bb(), "deg")
```

You can also use Galactic UVW velocities instead of proper motions and radial velocity by setting `uvw=True`:

```python
# From Galactic UVW velocities
o_uvw = Orbit(
    [150.0, 40.0, 1.5, -10.0, 20.0, 5.0],
    lb=True,
    uvw=True,
    ro=8.0,
    vo=220.0,
)
print("R =", o_uvw.R(), "kpc")
```

## From astropy SkyCoord

The most convenient initialization method uses astropy `SkyCoord`.

```python
from astropy.coordinates import SkyCoord

sc = SkyCoord(
    ra=20.0 * units.deg,
    dec=30.0 * units.deg,
    distance=2.0 * units.kpc,
    pm_ra_cosdec=3.5 * units.mas / units.yr,
    pm_dec=-1.2 * units.mas / units.yr,
    radial_velocity=15.0 * units.km / units.s,
)
o_sc = Orbit(sc)
print("R =", o_sc.R(), "kpc")
print("vT =", o_sc.vT(), "km/s")
```

<div class="alert alert-warning">

**Warning:** Subtleties when using SkyCoord:

- `galcen_distance` and `ro` are *not* interchangeable: `galcen_distance` is the
  distance between the Sun and the Galactic center, while `ro` is the projection of
  this distance onto the Galactic midplane.
- The astropy `Galactocentric` frame is right-handed, while galpy normally uses a
  left-handed frame. So the sign of the x-component of `galcen_v_sun` is the opposite
  of what it would be in `solarmotion`.

</div>

## The Sun's orbit

Calling `Orbit()` with no arguments returns the Sun's orbit.

```python
o_sun = Orbit()
print("Sun: R =", o_sun.R(), "kpc")
print("Sun: z =", o_sun.z(), "kpc")
```

## Turning physical units on and off

When orbits are initialized using `radec=True`, `lb=True`, or a `SkyCoord`, physical
scales `ro=` and `vo=` are automatically specified (using defaults of `ro=8` and `vo=220`).
All outputs will then be in physical units. You can toggle physical-unit output at any time.
If you want outputs in galpy's natural coordinates, turn physical output off:

```python
o_toggle = Orbit([1.0, 0.1, 1.0, 0.0, 0.05, 0.0], ro=8.0, vo=220.0)
print("Physical on:", o_toggle.R())

o_toggle.turn_physical_off()
print("Physical off:", o_toggle.R())

o_toggle.turn_physical_on()
print("Physical on again:", o_toggle.R())
```

## Example: initialize from SkyCoord, integrate, and plot

This example combines orbit initialization with integration and plotting.
For more details on orbit integration, see [Integration and Plotting](integration_and_plotting.ipynb).

```python
from astropy.coordinates import SkyCoord
from matplotlib import pyplot as plt

# Initialize from a SkyCoord (e.g., a nearby star)
sc = SkyCoord(
    ra=150.0 * units.deg,
    dec=20.0 * units.deg,
    distance=0.5 * units.kpc,
    pm_ra_cosdec=-5.0 * units.mas / units.yr,
    pm_dec=3.0 * units.mas / units.yr,
    radial_velocity=-20.0 * units.km / units.s,
)
o_example = Orbit(sc)
ts = numpy.linspace(0.0, 10.0, 10001)
o_example.integrate(ts, MWPotential2014)
o_example.plot()
```
